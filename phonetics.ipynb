```python
import plotly.express as px
```

```python
phoneme_df = pd.merge(gl_languages, data, on='Glottocode').drop(columns=['GlyphID', 'ISO6393', 'Allophones', 'Source'])
phoneme_df = pd.merge(gl_values, phoneme_df, left_on='Language_ID', right_on='ID').drop(columns=['ID_x', 'Code_ID', 'Comment', 'Source', 'codeReference'])
phoneme_df.head(5)
```

```python
phoneme_count_df = pd.concat([phoneme_df[['Phoneme', 'Name']].groupby(by=['Name'], as_index=False).count(), phoneme_df[['Latitude', 'Longitude', 'Name']].groupby(by=['Name'], as_index=False).mean().drop(columns='Name')], axis=1)
phoneme_count_df.head(5)
```

```python
fig2 = px.scatter_geo(phoneme_count_df, 
                     lat='Latitude', 
                     lon='Longitude',
                     hover_data=['Name', 'Phoneme'],
                     color = 'Name',
                     size = 'Phoneme',
                     projection="natural earth")
fig2.show()
```
